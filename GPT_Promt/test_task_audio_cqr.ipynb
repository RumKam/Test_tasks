## Расшифровка аудио

Для распознования текста будем использовать инструменты библиотеки SpeechRecognition и API Google Web Speech.

```python
# импорт библиотек
import speech_recognition as sr 
import os 
from pydub import AudioSegment
from pydub.silence import split_on_silence
from nltk import word_tokenize

import warnings
warnings.filterwarnings("ignore")
```

```python
# создадим объект для распознования
r = sr.Recognizer()

# зададим функцию для распознавания текста
def transcribe_audio(path):
    with sr.AudioFile(path) as source:
        audio_listened = r.record(source)
        text = r.recognize_google(audio_listened, language="ru-RU")
    return text

# функция для вывода большого текста
def large_transcribe_audio(path):
    sound = AudioSegment.from_file(path)  
    chunks = split_on_silence(sound,
        min_silence_len = 500,
        silence_thresh = sound.dBFS-14,
        keep_silence=500)
    folder_name = "audio-chunks"
    if not os.path.isdir(folder_name):
        os.mkdir(folder_name)
    whole_text = ""
    for i, audio_chunk in enumerate(chunks, start=1):
        chunk_filename = os.path.join(folder_name, f"chunk{i}.wav")
        audio_chunk.export(chunk_filename, format="wav")
        try:
            text = transcribe_audio(chunk_filename)
        except sr.UnknownValueError as e:
            print("неразборчиво")
        else:
            text = f"{text.capitalize()}. "
            print('реплика', ":", text)
            whole_text += text
    return whole_text
```

```python
#получим первый текст
path = "audio3.wav"
text1 = large_transcribe_audio(path)
```

```python
# получим второй текст
path = "implantcity.wav"
text2 = large_transcribe_audio(path)
```

```python
# получим третий текст
path = "whiteservice1.wav"
text3 = large_transcribe_audio(path)
```
